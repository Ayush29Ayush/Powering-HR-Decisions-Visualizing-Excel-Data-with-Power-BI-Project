# Data Analysis and Visualization

```python
df.head()
```

```python
cancelled_perc = df['is_canceled'].value_counts(normalize = True)
print(cancelled_perc)
```

```python
plt.figure(figsize=(5,4))
plt.title('Reservation Status Count')
plt.bar(['Not Cancelled','Cancelled'], df['is_canceled'].value_counts(), edgecolor = 'k', width=0.7)
# we cna observe that cancellation rate is more than half of booking rate. This is a serious problem
```

```python
# Now lets check individually, the cancellation plot in both the types of hotels
plt.figure(figsize = (8,4))
ax1= sns.countplot(x = 'hotel', hue = 'is_canceled', data = df, palette = 'Dark2')
plt.title('Reservation Status in Different Hotels', size = 20)
plt.xlabel('Hotel')
plt.ylabel('No. of Reservations')
plt.legend(['Not Cancelled', 'Cancelled'])
plt.show()
```

```python
# isse essa dataframe banao where hotels mein sirf resort hotels ho
# resort_hotel = df['hotel'] == 'Resort Hotel'
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel
```

```python
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
# here we can observe that city hotel's cancellation rate is way higher than the other
```

```python
# Now lets see day wise average daily rate. 
# To do se use GROUP_BY function like in SQL
# X = reservation date
# Y = avg of adr
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
resort_hotel
```

```python
# The code resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean() groups the resort_hotel DataFrame by reservation_status_date column and calculates the mean of the adr column for each group.

# The resulting DataFrame will have reservation_status_date as the index and adr column with the mean values. The double brackets around ['adr'] is used to return a DataFrame with a single column, instead of a pandas Series.
```

```python
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel
```

```python
plt.figure(figsize = (20,8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize = 30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
# here index is "reservation_status_date"
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()

# The spikes here can be because of high booking rates on weekends
```

```python
# Now lets check cancellation/month
df['month'] = df['reservation_status_date'].dt.month # month column nahi tha toh ek create kar diya from reservation_status_date
plt.figure(figsize = (16,8))
ax1 = sns.countplot(x = 'month', hue = 'is_canceled', data = df, palette = 'Set3')
# legend_labels,_ = ax1. get_legend_handles_labels()
# ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status per month', size = 20)
plt.xlabel('Month')
plt.ylabel('Number of Reservations')
plt.legend(['Not Cancelled', 'Cancelled'])
plt.show()
```

```python
# Now we can observe that when cancellation rates are low, reservation rates are high. But why is this happening?
# Is it because the price offered that month was high so users chose cancelling over reservation?
# Lets check Avg Daily Rate per month
```

```python
plt.figure(figsize=(16,8))
plt.title("Average Daily Rate per Month", fontsize=30)
sns.barplot(x='month', y='adr', data=df)
plt.show()
```

```python
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace = True)
cancelled_df_adr.sort_values('reservation_status_date', inplace = True)

not_cancelled_data = df[df['is_canceled'] == 0]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace = True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace = True)

plt.figure(figsize = (20,6))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'], label = 'not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'], label = 'cancelled')
plt.legend()
```
